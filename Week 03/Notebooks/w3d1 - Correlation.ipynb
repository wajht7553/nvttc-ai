# What is Correlation?

Correlation describes the relationship between two variables. It helps us understand whether and how strongly pairs of variables are related.

## Practical Example
Suppose we want to know if there is a relationship between students' study hours and their exam scores. By plotting these two variables, we can visually inspect if higher study hours tend to result in higher scores.

Let's see how to visualize and calculate correlation using Python.

```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate synthetic data: study hours vs exam scores
np.random.seed(42)
study_hours = np.random.normal(5, 2, 100)
exam_scores = 50 + study_hours * 8 + np.random.normal(0, 10, 100)

df = pd.DataFrame({'Study Hours': study_hours, 'Exam Score': exam_scores})

# Scatter plot to visualize correlation
plt.figure(figsize=(8,5))
sns.scatterplot(x='Study Hours', y='Exam Score', data=df)
plt.title('Study Hours vs Exam Score')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.show()
```

## 2. Correlation Coefficient

The **correlation coefficient** quantifies the strength and direction of a linear relationship between two variables. The most common is the Pearson correlation coefficient, which ranges from -1 (perfect negative) to +1 (perfect positive).

- **Positive correlation**: As one variable increases, the other tends to increase.
- **Negative correlation**: As one variable increases, the other tends to decrease.
- **Zero correlation**: No linear relationship.

### Calculating the Correlation Coefficient in Python

```python
# Calculate Pearson correlation coefficient
corr = df['Study Hours'].corr(df['Exam Score'])
print(f"Pearson correlation coefficient: {corr:.2f}")

# Interpretation
if corr > 0.7:
    print('Strong positive correlation')
elif corr > 0.3:
    print('Moderate positive correlation')
elif corr > 0:
    print('Weak positive correlation')
elif corr < -0.7:
    print('Strong negative correlation')
elif corr < -0.3:
    print('Moderate negative correlation')
elif corr < 0:
    print('Weak negative correlation')
else:
    print('No linear correlation')
```

## 3. Univariate, Bivariate, and Multivariate Plots

### Univariate Plots
Univariate plots show the distribution of a single variable. Common examples include histograms and boxplots.

### Bivariate Plots
Bivariate plots visualize the relationship between two variables. Scatter plots and line plots are typical examples.

### Multivariate Plots
Multivariate plots display relationships among three or more variables. Pair plots and heatmaps are useful for exploring complex datasets.

Let's see practical examples of each type.

```python
# Univariate Plot: Histogram of Study Hours
plt.figure(figsize=(7,4))
sns.histplot(df['Study Hours'], bins=15, kde=True)
plt.title('Distribution of Study Hours')
plt.xlabel('Study Hours')
plt.show()

# Bivariate Plot: Scatter plot already shown above

# Multivariate Plot: Pairplot
sns.pairplot(df)
plt.suptitle('Pairplot of Study Hours and Exam Score', y=1.02)
plt.show()
```

## 4. Introduction to Probability

Probability is a branch of mathematics that measures the likelihood of an event occurring. It is fundamental to statistics, data science, and machine learning.

- Probability values range from 0 (impossible event) to 1 (certain event).
- Used to model uncertainty and randomness in data.

### Practical Example: Coin Toss

Suppose we toss a fair coin. What is the probability of getting heads?

Let's simulate this in Python.

```python
# Simulate 1000 coin tosses
coin_tosses = np.random.choice(['Heads', 'Tails'], size=1000)
heads_count = np.sum(coin_tosses == 'Heads')
prob_heads = heads_count / len(coin_tosses)
print(f"Simulated probability of getting heads: {prob_heads:.2f}")
```
